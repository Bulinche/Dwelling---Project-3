```python
import pandas as pd
import requests
import json
```

```python
busStops = pd.read_csv('Bus - Stops - Final.csv')
```

```python
GoogleAPI='AIzaSyCJCoD8hXTCvE4Wswuu-Kvra5FjeRFJSxY'
```

```python
busStops.head()
```

```python
latlonTocity = f"https://maps.googleapis.com/maps/api/geocode/json?sensor=true&key={GoogleAPI}&latlng="
cityNames=[]

for index,row in busStops.iterrows():
    query = f"{latlonTocity}{row['stop_lat']},{row['stop_lon']}"
    result = requests.get(query).json()
    try:
        cityNames.append(result['results'][0]['address_components'][1]['long_name'])
    except:
        cityNames.append("NotFound")
        print(f"{index}: ( {row['stop_lat']} , {row['stop_lon']} ) not found")
#     print(query)
```
