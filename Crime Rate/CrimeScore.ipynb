```python
import pymongo

conn2 = 'mongodb://localhost:27017'
city_names = []
test = []
# Pass connection to the pymongo instance.
client2 = pymongo.MongoClient(conn2)

# Connect to a database. Will create one if not already available.
db2 = client2.Dwelling_db
cities = db2.city_lat_lon.find()
goldenC = []

for x in cities:
    city = x['City']
    latlng = f"{x['Latitude']},{x['Longitude']}"
    
    temp = {
        'city': city,
        'latlong' : latlng
    }
    goldenC.append(city)
    city_names.append(temp)
# test.sort(function (a,b){return a.City < b.City})


city_names = sorted(city_names, key = lambda i: i['city'])
city_names
```

```python
def regulators(val):
    print(val)
    if(val is None):
        return None
    elif(val>300):
        return float(0)
    elif(val>241):
        return float(100)
    elif(val>181):
        return float(200)
    elif(val>121):
        return float(300)
    elif(val>61):
        return float(400)
    else:
        return float(500)
```

```python
for i,r in temp.iterrows():
    temp.loc[i,'2005'] = regulators(r['2005'])
    temp.loc[i,'2006'] = regulators(r['2006'])
    temp.loc[i,'2007'] = regulators(r['2007'])
    temp.loc[i,'2008'] = regulators(r['2008'])
    temp.loc[i,'2009'] = regulators(r['2009'])
    temp.loc[i,'2010'] = regulators(r['2010'])
    temp.loc[i,'2011'] = regulators(r['2011'])
    temp.loc[i,'2012'] = regulators(r['2012'])
    temp.loc[i,'2013'] = regulators(r['2013'])
    temp.loc[i,'2014'] = regulators(r['2014'])
    temp.loc[i,'2015'] = regulators(r['2015'])
    temp.loc[i,'2016'] = regulators(r['2016'])
    temp.loc[i,'2017'] = regulators(r['2017'])
    temp.loc[i,'2018'] = regulators(r['2018'])
```
